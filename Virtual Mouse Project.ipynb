```python
import mediapipe as md
import cv2
import pyautogui as py
```

```python
cap = cv2.VideoCapture(0)
hand_detect = md.solutions.hands.Hands()
drawing_utils = md.solutions.drawing_utils
screen_width, screen_height = py.size()
index_y = 0

while True:
    _, frame = cap.read()
    frame = cv2.flip(frame, 1)
    frame_height, frame_width, _ = frame.shape
    
    
    
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    output = hand_detect.process(rgb_frame)
    hands = output.multi_hand_landmarks
    
    
    if hands:
        for i in hands:
            drawing_utils.draw_landmarks(frame, i)
            landmarks = i.landmark
            for id, landmark in enumerate(landmarks):
                x = int(landmark.x * frame_width)
                y = int(landmark.y * frame_height)
                
                if id == 8:
                    cv2.circle(img = frame, center = (x, y), radius = 10, color = (0, 255, 100))
                    index_x = screen_width / frame_width * x
                    index_y = screen_height / frame_height * y
                    py.moveTo(index_x, index_y)
                    
                if id == 4:
                    cv2.circle(img = frame, center = (x, y), radius = 10, color = (255, 150, 0))
                    thumb_x = screen_width / frame_width * x
                    thumb_y = screen_height / frame_height * y
                    
                    rick = int(abs(index_y - thumb_y))
                      
                    if rick < 22:
                        print("click screen ritwik", rick)

                    else:
                        print("out of need Ritwik!!!!!!!!!!!!!!!!", rick)
                    


                    if abs(index_y - thumb_y) < 22:
                        py.click()
                        py.sleep(3)
            
            
    cv2.imshow("virtual mouse", frame)
    cv2.waitKey(1)
```
